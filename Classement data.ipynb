```python
fig, axes = plt.subplots(5,5, figsize=(15,10), sharex='col')

for i in range(5):
    for j in range(4):
        indice = i*5+j
        prices = json.load(open(f"data/{i*5+j}.json"))
        axes[4-i, j].plot(prices[3_000:])
        
    
for i in range(5):
    indice = i*5+4
    axes[4-i, 0].set_ylabel(r"P=" + str(df.loc[indice, "P"]))
        
for j in range(5):
    indice = 4*5+j
    axes[4, j].set_xlabel(r"g_sur_l=" + str(df.loc[indice, "g_sur_l"]))
    
plt.legend()

plt.savefig("Phases du diagramme", dpi=600, bbox_inches='tight', pad_inches=0)
```
